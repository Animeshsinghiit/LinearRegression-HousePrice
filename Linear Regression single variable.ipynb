```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
housep=pd.concat([data.LotArea,data.SalePrice], axis="columns")
```

```python
housep.head()
```

```python
%matplotlib inline
plt.scatter(housep.LotArea,housep.SalePrice, color="red",marker="*")
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(housep[["LotArea"]],housep.SalePrice)
```

```python
reg.predict([[3245]])
```

```python
tdata=pd.read_csv("test.csv")
tdata.head()
```

```python
test=pd.DataFrame(tdata.LotArea)
t1=test.head()
```

```python
t1
```

```python
# test["SalePrice"]=""
pred=[]
for i in test.LotArea:
    pred.append(reg.predict([[i]]))
```

```python
predD=pd.DataFrame(pred)
```

```python
predD.head()
```

```python
final=pd.concat([test,predD],axis="columns")
```
